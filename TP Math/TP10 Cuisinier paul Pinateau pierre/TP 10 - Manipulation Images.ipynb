<p>En utilisant la commande <span style="font-family: comic sans ms,sans-serif;"><strong>A.solve_right(B)</strong></span> ou <span style="font-family: comic sans ms,sans-serif;"><strong>A\b</strong></span> r&eacute;soudre les syst&egrave;mes lineaires suivants et v&eacute;rifier vos solutions:</p>
<p><span style="text-decoration: underline;"><br /></span></p>
$$(S_1)\left\{\begin{array}{lcl}x+y-z & = & 2 \\ 2x-y+z & = & 1 \\ 4x-y+3z & = & 3\end{array}\right. \qquad (S_2)\left\{\begin{array}{lcl}x-2y+z-4t & = & 1 \\ x+3y+7z+2t & = & 2 \\ x-12y-11z-16t & = & 3 \end{array}\right.$$

```python
A=matrix([[1,1,-1],[2,-1,1], [4,-1,3]])
B=matrix([[2],[1],[3]])

X = A.solve_right(B)
print("S1")
show(X)

A=matrix([[1,-2,1,-4],[1,3,7,2], [1,-12,-11,-16]])
B=matrix([[1],[2],[3]])
print("S2")
try:
    X = A.solve_right(B)
    show(X)
except Exception as e:
    print(e)
```

<b>Traitement mathématiques des images numérique</b> <br>

Vous avez maintenant bien compris que l'algèbre linéaire et l'analyse matricielle vont nous permettre de manipuler des images.<br>

<b>Acquisition d’une image (= numérisation)</b> : dans la figure ci dessous on a à gauche une image réelle et à droite
sa version numérisée. La numérisation d’une image réelle comporte une perte d’information due d’une part à
l’échantillonnage (procédé de discrétisation spatiale d’une image consistant à projeter sur une grille régulière, i.e.
en un nombre fini de points, une image analogique continue en lui associant une valeur unique), d’autre part à la
quantification (nombre finis de nuances = limitation du nombre de valeurs différentes que peut prendre un point ).

<b> Les pixels d’une image </b>: une image numérique en niveaux de gris (<i>grayscale</i> image en anglais) est un tableau de
valeurs $A$. Chaque case de ce tableau, qui stocke une valeur $a_{ij}$ , se nomme un pixel (<i>PICTure ELement</i>).
En notant $n$ le nombre de lignes et $p$ le nombre de colonnes du tableau, on manipule ainsi un tableau de $n \times p$ pixels.
Les valeurs des pixels sont enregistrées dans l’ordinateur ou l’appareil photo numérique sous forme de nombres
entiers entre 0 et 255, ce qui fait 256 valeurs possibles pour chaque pixel. La valeur 0 correspond au noir et la valeur
255 correspond au blanc. Les valeurs intermédiaires correspondent à des niveaux de gris allant du noir au blanc.
On peut définir une fonction comme suit :
    $$g:[1:n]\times[1:p]\rightarrow [0:255]$$
    $$(i,j)\rightarrow g(i,j)=a_{ij}$$

<b>La taille d’une image</b> : la taille d’une image est le nombre de pixel. 
Les dimensions d’une image sont la largeur (=nombre de colonnes du tableau) et la hauteur (=nombre de lignes du tableau).

<b>La résolution d’une image</b> : la résolution d’une image est le nombre de pixel par unité de longueur. 

En générale, on utilise des “pixel par pouce” ou “point par pouce” (ppp), on anglais on dit <i>dot per inch</i> (dpi) : $n$ dpi = $n$ pixel pour un puce = $n$ pixel pour 2.54 cm.
